```python
'''
Standard Resolutions: 
1080p: 1920x1080
720p: 1280x720
480p: 854x480
360p: 640x360
240p: 426x240
'''
# Scale Down or Up to our RESOLUTION
# Reduce or Increase the COLOUR Channels*
# Extract a fixed amount of FRAMES

import cv2
import os

SAVE_PATH = 'D:/Ingested'
TARGET_PATH = 'D:/RawData'
FRAME_RATE = 10
RESOLUTION_WIDTH = 640
RESOLUTION_HEIGHT = 360

for root,dirs,files in os.walk(TARGET_PATH):
    for file in files:
        if ('.MOV' in file):
            path = os.path.join(root, file)
            capture = cv2.VideoCapture(path)
            fourcc = cv2.VideoWriter_fourcc(*'XVID')
            out = cv2.VideoWriter(SAVE_PATH + '/' + file+'_formatted.mp4v',fourcc, FRAME_RATE, (RESOLUTION_WIDTH,RESOLUTION_HEIGHT))
            while True:
                isTrue, frame = capture.read()
                if isTrue == True:
                    b = cv2.resize(frame,(RESOLUTION_WIDTH,RESOLUTION_HEIGHT),fx=0,fy=0, interpolation = cv2.INTER_CUBIC)
                    out.write(b)
                else:
                    break
            capture.release()
            out.release()
            cv2.destroyAllWindows()

from datetime import datetime

origin="00:00:00"          #the origin
start="00:00:00"           #specify start time in hh:mm:ss
end = "00:00:01"           #specify end time in hh:mm:ss

origintime=datetime.strptime(origin,"%H:%M:%S") #origin 
starttime=datetime.strptime(start,"%H:%M:%S")  #start time
endtime=datetime.strptime(end,"%H:%M:%S")      #end time

startframe=FRAME_RATE*(starttime-origintime).total_seconds()  #get the start frame
endframe=FRAME_RATE*(endtime-origintime).total_seconds()  #get the end frame

counter =1 #set counter
for root,dirs,files in os.walk(SAVE_PATH):
    for file in files:
        if ('.MOV' in file):
            path = os.path.join(root, file)
            capture = cv2.VideoCapture(path)
            fourcc = cv2.VideoWriter_fourcc(*'XVID')
            out = cv2.VideoWriter(SAVE_PATH + '/' + file+'_&trimmed.mp4v',fourcc, FRAME_RATE, (RESOLUTION_WIDTH,RESOLUTION_HEIGHT))
            counter = 1
            while True:
                isTrue, frame = capture.read()
                if isTrue == True:
                    if counter>=startframe and counter<=endframe: 
                        b = cv2.resize(frame,(RESOLUTION_WIDTH,RESOLUTION_HEIGHT),fx=0,fy=0, interpolation = cv2.INTER_CUBIC)
                        out.write(b)
                    counter += 1  #increase counter
                else:
                    break
            capture.release()
            out.release()
            cv2.destroyAllWindows()
```
